```python
df = pd.read_csv('insurance.csv')

colors = {
    'male':'blue',
    'female':'red'}

df['colors'] = [colors[x] for x in df['sex']]
df
source = ColumnDataSource(data=df)

p=figure(
    title ='plot bmi vs charges',
    x_axis_label ='bmi',
    y_axis_label ='charges'

)

p.circle(
    source = source,
    x='bmi',
    y='charges',
    colors = colors
)



show(p)
```

```python
from bokeh.sampledata.iris import flowers  # flowers is a Pandas dataframe
```

```python
colors ={'setosa' : 'blue',
         'versicolor': 'green',
         'virginica' : 'red'}

flowers['colors'] = [colors[x] for x in flowers ['species']]
source1 = ColumnDataSource(data=flowers)

p1 = figure(
    title ='Plot of different species',
    y_axis_label ='petal_length',
    x_axis_label = 'petal_width'  
)
p1.xaxis.axis_label = 'petal_length'
p1.yaxis.axis_label = 'petal_width'

p1.circle(
    source = source1,
    x ='petal_length',
    y = 'petal_width',
    color = 'colors',
    size=10,
    fill_alpha=0.2,
    legend_field='species'

)

p1.legend.location='top_left'
#p1.add_layout(p1.legend[0], 'right')
p1.add_layout(p1.legend[0], 'left')
output_file('xx.html')
show(p1)
```
